1\. Create a plot with graphs of two functions: $y=e^{-x^2}$ and $y=3x^2 e^{-x^2}$. Add x and y axes labels and the legend. Turn on the grid. Customize curve colors and styles.

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def f1(x):
    return np.exp(-pow(x, 2))

def f2(x):
    return 3 * pow(x, 2) * np.exp(-pow(x, 2))
```

```python
N = 100

x1 = [x for x in np.arange(0, 1, 1/N)]
y1 = [f1(x) for x in x1]

x2 = [x for x in np.arange(0, 1, 1/N)]
y2 = [f2(x) for x in x2]
```

```python
plt.plot(x1, y1, label = "y=e^(âˆ’x^2)", color="b")
plt.plot(x2, y2, label = "y=3x2e^(-x^2)", color="r")
plt.legend()
plt.grid()
plt.show()
```

1\. Create 3D surface plot of a function $z=x^2-y^2$. Add axes labels and color bar.

```python
from matplotlib import cm
from matplotlib.ticker import LinearLocator
```

```python
fig, ax = plt.subplots(figsize=(12, 6), subplot_kw={"projection": "3d"})

X = np.arange(-1, 1, 0.05)
Y = np.arange(-1, 1, 0.05)
X, Y = np.meshgrid(X, Y)
Z = X**2 - Y**2

surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False)

fig.colorbar(surf, shrink=0.5, aspect=5)

front_size = 15
ax.set_xlabel('$X$', fontsize=front_size)
ax.set_ylabel('$Y$', fontsize=front_size)
ax.set_zlabel('$Z$', fontsize=front_size)

plt.show()
```
