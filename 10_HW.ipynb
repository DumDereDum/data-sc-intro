1\. Download the file ""redundant1.csv"" from the repository "https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/". 
Apply PCA to reveal its number of essential features. Compute the reduced dataset and create a scatter plot for it.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv("data/redundant1.csv",
                   header = 0,
                   names=['1','2','3','4','5','6','7'])
Y = data.apply(lambda x: x-x.mean())
Z = preprocessing.normalize(Y, axis=0)
data = pd.DataFrame(Z)
```

```python
data.describe()
```

```python
pca_breast = PCA(n_components=2)
principalComponents_breast = pca_breast.fit_transform(data)
principal_cifar_Df = pd.DataFrame(
    data = principalComponents_breast,
    columns = ['principal component 1', 'principal component 2'])
```

```python
plt.figure(figsize=(16,10))
sns.scatterplot(
    x="principal component 1", y="principal component 2",
    palette=sns.color_palette("hls", 10),
    data=principal_cifar_Df,
    legend="full",
    alpha=0.3
)
```

2\. Download the file ""redundant2.csv"" from the repository "https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/". 
Apply NMF and extract two its main features. Plot scatter plot for them.

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.decomposition import NMF
```

```python
data = pd.read_csv("data/redundant2.csv",
                    header = 0,
                    names=[str(i) for i in range(100)])
data.describe()
```

```python
nmf = NMF(n_components=2,
          init='random',
          random_state=0,
          max_iter=10000)
W = nmf.fit_transform(data)
fig, axs = plt.subplots(1, 1,
                        figsize=(10, 4),
                        sharex=True)
axs.scatter(W[:, 1], W[:, 0])
```
