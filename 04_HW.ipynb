```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data/bank_customers.csv")
```

```python
data.head()
```

```python
mean   = data['CustAge'].mean()
median = data['CustAge'].median()
mode   = data['CustAge'].mode()

plt.figure(figsize=(10, 5))
plt.hist(data['CustAge'], bins=30)

plt.axvline(x = mean,    color = 'r', label = 'mean')
plt.axvline(x = median,  color = 'g', label = 'median')
plt.axvline(x = mode[0], color = 'y', label = 'mode')

plt.legend()
plt.grid()
plt.show()
```

1\. File `"bank_customers.csv"` that you can find at `"https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/"` describes bank clinets. It contains three columns: customer age (`CustAge`), period of relationship with the bank in months (`Months`), and credit limit (`CredLim`). Plot a histogram for the column `CustAge` and compute its standard deviation and the interquartile range. Show them on the plot.

```python
def my_quantile(data, q):
    """
    The quantile, q in (0,1)
    """
    inx = int(round(q * len(data)))
    return sorted(data)[inx]
def my_interquart(data):
    return my_quantile(data, 0.75) - my_quantile(data, 0.25)
```

```python
column = 'CustAge'
mean   = data[column].mean() 
median = data[column].median()
std    = data[column].std()
iqr    = my_interquart(data[column])
```

```python
plt.figure(figsize=(10, 5))
plt.hist(data[column], bins=30)

plt.axvline(x = mean, color = 'r', label = 'mean')
plt.axvline(x = mean + std,  color = 'orangered', label = 'std')
plt.axvline(x = mean - std,  color = 'orangered', label = 'std')

plt.axvline(x = median,  color = 'g', label = 'median')
plt.axvline(x = median+iqr,  color = 'lime', label = 'iqr')
plt.axvline(x = median-iqr,  color = 'lime', label = 'iqr')

plt.legend()
plt.grid()
plt.show()
```
